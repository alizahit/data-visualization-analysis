```python
df=pd.read_csv(r"C:\Users\aliza\data_visualization\Fish.csv")
df
```

```python
df_mean=df.groupby('Species').mean()
```

```python
df_bream=df.loc[df['Species']=='Bream']
df_parkki=df.loc[df['Species']=='Parkki']
df_perch=df.loc[df['Species']=='Perch']
df_pike=df.loc[df['Species']=='Pike']
df_roach=df.loc[df['Species']=='Roach']
df_smelt=df.loc[df['Species']=='Smelt']
df_whitefish=df.loc[df['Species']=='Whitefish']
```

```python
#df_bream.head(5)
```

```python
fig, ax= plt.subplots()
ax.set_title('Weight')

a=[df_bream['Weight'],df_parkki['Weight'],df_perch['Weight'],
  df_pike['Weight'],df_roach['Weight']]
ax1=plt.violinplot(a)
for ax11 in ax1['bodies']:
    ax11.set_facecolor('red')

b=['bream','parkki','perch','pike','roach']
ax.set_xticks(range(1,len(b)+1))
ax.set_xticklabels(b)
plt.savefig('Weight violin.png')
```

```python
tur=[df_bream,df_parkki,df_perch,df_pike,df_roach,df_smelt,df_whitefish]
isaret=["o","s","^","x",".","d","_"]
turr=['bream',"parkki","perch","pike","roach","smelt","whitefish"]

for i,ii in zip(tur,isaret):
    plt.scatter(i['Height'],i['Width'],marker=ii,alpha=1)
    

plt.legend(turr,)
plt.xlabel('Height')
plt.ylabel('Width')

plt.savefig('Species scatter.png')
```
